```python
import matplotlib.pyplot as plt

runs = [1,2,3,4,5]

# values are in ms
lambda_data = [
    [397, 90, 15, 13, 8], 
    [244, 66, 10, 9, 8], 
    [303, 27, 27, 12, 11], 
    [290, 43, 21, 15, 9], 
    [260, 19, 12, 10, 9]
]


# values are in s
DB_data = [
    [0.239779949,
    0.039309978,
    0.039899826,
    0.039960146,
    0.040019989],

    [0.209889889,
    0.038810015,
    0.040149927,
    0.039729834,
    0.040070057],

    [0.220210075,
    0.038350105,
    0.03990984,
    0.039979935,
    0.039969921],

    [0.217309952,
    0.058520079,
    0.039769888,
    0.03986001,
    0.039930105],

    [0.216030121,
    0.039690018,
    0.03993988,
    0.019939899,
    0.040019989]
]


def average(arr):
    res = []

    for i in range(len(arr)):
        S = 0
        for j in range(len(arr[i])):
            S += arr[j][i]
        res.append(S / len(arr[i]))
    
    return res

def multiply(arr, m):
    return [elem * m for elem in arr]

plt.plot(runs, average(lambda_data))
plt.title('AWS Lambda Delays (Averaged over 5 runs)')
plt.xlabel('Consecutive Invocation #')
plt.ylabel('Delay (ms)')
plt.xticks([1,2,3,4,5])
plt.show()
```

```python
plt.plot(runs, multiply(average(DB_data), 1000)) # multiply each by 1000 to convert from s to ms
plt.title('AWS DynamoDB Delays (Averaged over 5 runs)')
plt.xlabel('Consecutive Invocation #')
plt.ylabel('Delay (ms)')
plt.xticks([1,2,3,4,5])
plt.show()
```
